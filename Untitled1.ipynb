```python
text = [i for i in df['text']]
```

```python
label = np.array(label)
text = np.array(text)
```

```python
text.shape
```

```python
maxlen = max([len(i) for i in text])
print(maxlen)
```

```python
from warnings import filters
Tok = Tokenizer(filters='!#$%^&*()_+{}:">?<~')
```

```python
Tok.fit_on_texts(text)
```

```python
print(len(Tok.word_counts))
```

```python
print(Tok.word_index)
Seq=Tok.texts_to_sequences(text)
padded = pad_sequences(Seq,padding='post',maxlen=maxlen,truncating='post')
```

```python
trainig_texts = padded[:5000]
trainig_labels = label[:5000]
test_texts = padded[5000:]
test_labels = label[5000:]
```

```python
model = keras.Sequential([
    keras.layers.Embedding(80785,32,input_length=maxlen),
    keras.layers.Bidirectional(keras.layers.LSTM(64)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics=['accuracy']

)
```

```python
h = model.fit(trainig_texts,trainig_labels,epochs=5)
```

```python
model.evaluate(test_texts,test_labels)
```

```python
sample_email = ['Hey Shyam prasad,You are Eligible Internship & Fresher Job through Zuno Register (Last Date: 27th Aug)']
sample_email = np.array(sample_email)
ts = Tok.texts_to_sequences(sample_email)

pad_te = pad_sequences(ts,padding='post',maxlen=maxlen,truncating='post')
```

```python
p = model.predict(pad_te)
p = p[0]
```

```python
if p>=0.5:
  print("message is spam")
else:
  print("Not spam")
```
